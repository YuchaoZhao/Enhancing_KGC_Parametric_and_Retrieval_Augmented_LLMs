```python
import numpy as np
import pandas as pd
import os
import json
```

```python
def read_triples(filepath:str):
    triples = []
    with open(filepath) as f:
        for line in f:
            data_instance = json.loads(line)
            triples.append(data_instance)
    return triples
```

```python
def write_modification_file(output_path:str, results:list):
    with open(output_path, 'w') as f:
        for result in results:
            json.dump(result, f)  # JSON encode each element and write it to the file
            f.write('\n')
```

```python
# Method 8 - XGBoost   49.79% (47.62%)
import pandas as pd
from sklearn.preprocessing import OneHotEncoder, StandardScaler, LabelEncoder
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Load the train-set and test-set
train_df = pd.read_csv('train_data.csv')
test_df = pd.read_csv('test_data.csv')

# Preprocess the data
cat_cols = ['prop', 'class']
num_cols = ['pop']
enc = OneHotEncoder(drop='first', sparse=False, handle_unknown='infrequent_if_exist')
scaler = StandardScaler()
train_data = pd.concat([pd.DataFrame(enc.fit_transform(train_df[cat_cols])),
                        pd.DataFrame(scaler.fit_transform(train_df[num_cols]), columns=num_cols)],
                       axis=1)

test_data = pd.concat([pd.DataFrame(enc.transform(test_df[cat_cols])),
                       pd.DataFrame(scaler.transform(test_df[num_cols]), columns=num_cols)],
                      axis=1)

# Convert column names to strings
train_data.columns = train_data.columns.astype(str)
test_data.columns = test_data.columns.astype(str)

# Map labels to binary integers
label_encoder = LabelEncoder()
train_labels = label_encoder.fit_transform(train_df['label'])

# Split the train-data into training and validation sets
train_X, val_X, train_y, val_y = train_test_split(train_data, train_labels, test_size=0.2, random_state=42)

# Choose models
# version8
model = XGBClassifier(random_state=42)

# Train the model
model.fit(train_X, train_y)

# Evaluate the model on the validation set
val_preds = model.predict(val_X)
val_acc = accuracy_score(val_y, val_preds)
print('Validation accuracy:', val_acc)

# Predict the labels for the test-set
test_preds = model.predict(test_data)

# Map the predicted labels back to the original class labels
predicted_labels = label_encoder.inverse_transform(test_preds)

# Save the predictions in a CSV file
test_df['label'] = predicted_labels
test_df.to_csv('kamel_20_with_label.csv', index=False)
```
